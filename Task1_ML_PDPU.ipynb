```
import pandas as pd
import datetime
import dateparser
```

```
'''
•	Load the data.
•	Read first 5 rows to get headers
'''
data = pd.read_csv("household_power_consumption.txt", sep =';')
data
```

```
data.head(5)
```

```
data['Date'] + ' ' + data['Time']
```

```
'''
•	Converting Date and Time variables to Date/Time format    
'''
data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])
```

```
data
```

```
'''
•	Read 2900 rows that contain information on 2007-02-01 and 2007-02-02
•	Subset the loaded data for 2007-02-01 and 2007-02-02
'''
data['datetime'] = pd.to_datetime(pd.Series(data.datetime))
print('1 done')
data['month'] = pd.DatetimeIndex(data['datetime']).month
data['year'] = pd.DatetimeIndex(data['datetime']).year
data['date'] = pd.DatetimeIndex(data['datetime']).day

data.head()
```

```
# on 2007-02-01 and 2007-02-02
data[data['year'] == 2007][data['month'] == 2][data['date'] < 3]
```

```
data["Sub_metering_1"] = pd.to_numeric(data["Sub_metering_1"],errors='coerce')
data["Sub_metering_2"] = pd.to_numeric(data["Sub_metering_2"],errors='coerce')
data["Global_active_power"] = pd.to_numeric(data["Global_active_power"],errors='coerce')
```

```
print (data.dtypes)
```

```
'''
•	Histogram of Global active power consumption
'''
import matplotlib.pyplot as plt
plt.suptitle('Histogram of Global active power consumption')
plt.style.use('ggplot')
plt.hist(data["Global_active_power"], bins=10)
plt.show()
```

```
'''
•	Global active power consumption over time
'''
plt.suptitle('Global active power consumption over time')
### Line Plot
plt.figure(figsize=(19,4))   ### (width,height)
plt.title("Line Plot Graph")
plt.xlabel("Dates")
plt.ylabel("Global active power")
# plt.xlim([1,10])
plt.plot(data["datetime"],data["Global_active_power"],label="Line Graph",color='purple')
plt.legend(loc=2)
plt.show()
```

```
### Line Plot
plt.figure(figsize=(20,8))   ### (width,height)
plt.title('Energy Sub Metering')
plt.xlabel("Dates")
plt.ylabel("Energy Sub Metering")
plt.plot(data["datetime"],data["Sub_metering_1"],label="Sub metering 1",color='purple',lw=1)
plt.plot(data["datetime"],data["Sub_metering_2"],label="Sub metering 2",color='blue',lw=1)
plt.plot(data["datetime"],data["Sub_metering_3"],label="Sub metering 3",color='red',lw=1)
plt.legend(loc=2)
plt.show()
```
