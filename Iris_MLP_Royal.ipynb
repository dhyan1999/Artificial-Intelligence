```
import numpy as np
import pandas as pd
from keras.utils import to_categorical
```

```
dataset = pd.read_csv('/Iris.csv')
dataset.head()
```

```
x = dataset.iloc[:,1:5].values
y = dataset.iloc[:,5].values

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y1 = encoder.fit_transform(y)

Y = pd.get_dummies(y1).values
print(Y)
```

```
#### Split the data into training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,y1train,y1test = train_test_split(x,Y,test_size=0.2,random_state=0)
```

```
xtest
```

```
from keras.models import Sequential  ### For sequential input
from keras.layers import Dense
from keras.optimizers import SGD , Adam ### Optimizers

model = Sequential()

### Now to add layers to our model
### 10 = no of neurons
### 4 neurons is connected to 10 neurons in the next sequential Layer

model.add(Dense(10, input_shape=(4,), activation = 'sigmoid'))

### Adding 1 more layer
model.add(Dense(8, activation='sigmoid'))

### Adding 1 more layer
model.add(Dense(6, activation='sigmoid'))

### Adding Output layer
model.add(Dense(3, activation='softmax'))

### Compile the model
model.compile(Adam(lr = 0.01), 'categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```
model.fit(xtrain,y1train,epochs = 100)
y_pred = model.predict(xtest)
```
